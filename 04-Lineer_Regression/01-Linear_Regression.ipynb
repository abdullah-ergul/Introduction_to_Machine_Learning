# Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Experience-Salary_Dataset.csv")
```

```python
df
```

```python
df = pd.read_csv("Experience-Salary_Dataset.csv", sep=";")
```

```python
df
```

```python
plt.scatter(df.Experience,df.Salary)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

```python
x = df.Experience.values
```

```python
x.shape
```

```python
x = df.Experience.values.reshape(-1,1)
y = df.Salary.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x,y)
```

```python
b0 = linear_reg.predict([[0]])
```

```python
b0 = linear_reg.intercept_
print(b0)
```

```python
b1 = linear_reg.coef_
print(b1)
```

**General Formulla => Y = BO+B1*X**

```python
new_salary = 1663+1138*11
print(new_salary)
```

```python
b11 = linear_reg.predict([[11]])
print(b11)
```

```python
y_head = linear_reg.predict(x)
plt.plot(x,y_head,color="red")
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.metrics import r2_score
print("r square score", r2_score(y,y_head))
```
